üöÄ Informe Final del An√°lisis de Evasi√≥n de Clientes (Churn) en TelecomX_LATAM üìä
üîπ Introducci√≥n
El Churn de clientes, o la evasi√≥n, es un problema cr√≠tico para las empresas de telecomunicaciones como TelecomX_LATAM. La retenci√≥n de clientes existentes es a menudo m√°s rentable que la adquisici√≥n de nuevos. El objetivo de este an√°lisis es explorar un conjunto de datos de clientes, identificar los factores clave que influyen en la decisi√≥n de un cliente de darse de baja del servicio y, finalmente, proponer estrategias para mitigar este problema.

üîπ Limpieza y Tratamiento de Datos
El proceso de preparaci√≥n de datos fue crucial para asegurar la calidad del an√°lisis. Se llevaron a cabo las siguientes etapas:

Extracci√≥n y normalizaci√≥n: Se importaron los datos en formato JSON y se normalizaron para crear una estructura tabular (DataFrame) adecuada para el an√°lisis.

Eliminaci√≥n de datos irrelevantes: Se descartaron columnas como customerID y account_Charges_Total por ser identificadores √∫nicos o informaci√≥n redundante que no aportaba valor predictivo.

Manejo de valores ausentes y duplicados: Se verific√≥ la presencia de valores nulos, y se identificaron y eliminaron 27 filas duplicadas para evitar sesgos en el an√°lisis.

Estandarizaci√≥n de categor√≠as: Se unificaron las categor√≠as en columnas como phone_MultipleLines e internet_InternetService para garantizar la coherencia. Por ejemplo, "No phone service" y "No internet service" se reemplazaron por "No" o "No internet service" respectivamente, y en la columna Churn se trataron los valores vac√≠os como nulos (nan).

Creaci√≥n de nuevas variables: Se cre√≥ la columna Cuentas_Diarias dividiendo el Pago_Mensual por 30, lo que permiti√≥ un an√°lisis m√°s detallado del gasto promedio por d√≠a.

Codificaci√≥n de variables: Se transformaron las variables categ√≥ricas de tipo "S√≠/No" a valores binarios (1/0) y se renombraron todas las columnas a nombres en espa√±ol para mejorar la claridad y la legibilidad del conjunto de datos.

üîπ An√°lisis Exploratorio de Datos
A trav√©s del an√°lisis exploratorio, se identificaron varios patrones y correlaciones clave:

Distribuci√≥n del Churn: El 26.5% de los clientes en el conjunto de datos se dieron de baja. Esto subraya la importancia del problema y la necesidad de intervenir.

Variables categ√≥ricas y Churn:

Tipo de contrato: Los clientes con contratos mensuales (Month-to-month) tienen una tasa de evasi√≥n notablemente alta, con un 42.7%. En contraste, los contratos de uno y dos a√±os tienen tasas de churn muy bajas (11.3% y 2.8%, respectivamente), lo que demuestra que los contratos a largo plazo son un indicador de alta fidelidad.

Servicio de internet: El servicio de fibra √≥ptica (Fiber optic) muestra una tasa de churn significativamente mayor (41.8%) en comparaci√≥n con el servicio DSL (18.9%) y los clientes sin internet (7.2%). Esto podr√≠a indicar problemas de calidad de servicio, precios o insatisfacci√≥n espec√≠fica con la fibra √≥ptica.

M√©todo de pago: El pago electr√≥nico (Electronic check) est√° fuertemente asociado con una tasa de evasi√≥n del 45.1%. Los m√©todos de pago autom√°ticos (tarjeta de cr√©dito y transferencia bancaria) muestran tasas mucho m√°s bajas, lo que sugiere que la automatizaci√≥n del pago reduce la fricci√≥n y el riesgo de churn.

Facturaci√≥n: Los clientes que optan por la facturaci√≥n sin papel (Facturacion_Sin_Papel) tienen m√°s del doble de riesgo de churn (33.4%) que aquellos que reciben facturas en papel (16.3%).

Variables num√©ricas y Churn:

Antig√ºedad del cliente (Meses_Contrato): Los clientes que se dan de baja tienen una antig√ºedad promedio de solo 18 meses, significativamente menor que los clientes que se quedan (37 meses). Esto indica que los clientes nuevos son los que tienen mayor riesgo de irse.

Pago mensual (Pago_Mensual): Los clientes que se van tienen un pago mensual promedio m√°s alto ($74.6) que los que se quedan ($61.4). Esto podr√≠a sugerir que los clientes con facturas m√°s altas son m√°s sensibles al precio o perciben un menor valor por su dinero.

Servicios adicionales: La falta de servicios de seguridad en l√≠nea, respaldo, protecci√≥n de dispositivos y soporte t√©cnico (Seguridad_Online, Backup_Online, Proteccion_Dispositivo, Soporte_Tecnico) est√° relacionada con una mayor propensi√≥n a la evasi√≥n. Esto sugiere que estos servicios crean un "efecto pegamento" que mantiene a los clientes.

üîπ Conclusiones e Insights
El an√°lisis ha revelado que la evasi√≥n no es un fen√≥meno aleatorio, sino que est√° impulsado por factores espec√≠ficos que pueden ser identificados y abordados. Los clientes con el mayor riesgo de churn son aquellos que:

Tienen contratos mensuales.

Utilizan el servicio de fibra √≥ptica.

Pagan con cheques electr√≥nicos.

Son clientes relativamente nuevos.

Tienen facturas mensuales altas.

No han contratado servicios de valor agregado como seguridad o soporte t√©cnico.

No se encontr√≥ una correlaci√≥n significativa entre el g√©nero del cliente y la tasa de evasi√≥n.

üîπ Recomendaciones
Basado en los hallazgos, se proponen las siguientes estrategias para reducir la evasi√≥n de clientes:

Fidelizaci√≥n de nuevos clientes: Implementar programas de bienvenida y seguimiento proactivo para clientes con menos de 12 meses de antig√ºedad. Ofrecer incentivos o promociones exclusivas para motivar la renovaci√≥n de contratos anuales.

Transici√≥n a contratos a largo plazo: Promover activamente la migraci√≥n de clientes con contratos mensuales a contratos anuales o bienales, ofreciendo descuentos o beneficios adicionales.

Optimizaci√≥n del servicio de fibra √≥ptica: Investigar y mejorar la calidad del servicio de fibra √≥ptica para abordar las posibles causas de insatisfacci√≥n. Esto podr√≠a incluir mejoras en la red, atenci√≥n al cliente especializada o paquetes de servicio m√°s competitivos.

Promoci√≥n de servicios de valor agregado: Incentivar la adopci√≥n de servicios como seguridad en l√≠nea y soporte t√©cnico. Estos servicios no solo aumentan los ingresos, sino que tambi√©n mejoran la experiencia del cliente y la retenci√≥n.

Incentivar m√©todos de pago autom√°ticos: Promover activamente el uso de tarjetas de cr√©dito o transferencias bancarias autom√°ticas, ofreciendo un peque√±o descuento o beneficio a los clientes que migren de los cheques electr√≥nicos.
