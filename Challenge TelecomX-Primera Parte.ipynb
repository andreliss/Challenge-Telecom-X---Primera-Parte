🚀 Informe Final del Análisis de Evasión de Clientes (Churn) en TelecomX_LATAM 📊
🔹 Introducción
El Churn de clientes, o la evasión, es un problema crítico para las empresas de telecomunicaciones como TelecomX_LATAM. La retención de clientes existentes es a menudo más rentable que la adquisición de nuevos. El objetivo de este análisis es explorar un conjunto de datos de clientes, identificar los factores clave que influyen en la decisión de un cliente de darse de baja del servicio y, finalmente, proponer estrategias para mitigar este problema.

🔹 Limpieza y Tratamiento de Datos
El proceso de preparación de datos fue crucial para asegurar la calidad del análisis. Se llevaron a cabo las siguientes etapas:

Extracción y normalización: Se importaron los datos en formato JSON y se normalizaron para crear una estructura tabular (DataFrame) adecuada para el análisis.

Eliminación de datos irrelevantes: Se descartaron columnas como customerID y account_Charges_Total por ser identificadores únicos o información redundante que no aportaba valor predictivo.

Manejo de valores ausentes y duplicados: Se verificó la presencia de valores nulos, y se identificaron y eliminaron 27 filas duplicadas para evitar sesgos en el análisis.

Estandarización de categorías: Se unificaron las categorías en columnas como phone_MultipleLines e internet_InternetService para garantizar la coherencia. Por ejemplo, "No phone service" y "No internet service" se reemplazaron por "No" o "No internet service" respectivamente, y en la columna Churn se trataron los valores vacíos como nulos (nan).

Creación de nuevas variables: Se creó la columna Cuentas_Diarias dividiendo el Pago_Mensual por 30, lo que permitió un análisis más detallado del gasto promedio por día.

Codificación de variables: Se transformaron las variables categóricas de tipo "Sí/No" a valores binarios (1/0) y se renombraron todas las columnas a nombres en español para mejorar la claridad y la legibilidad del conjunto de datos.

🔹 Análisis Exploratorio de Datos
A través del análisis exploratorio, se identificaron varios patrones y correlaciones clave:

Distribución del Churn: El 26.5% de los clientes en el conjunto de datos se dieron de baja. Esto subraya la importancia del problema y la necesidad de intervenir.

Variables categóricas y Churn:

Tipo de contrato: Los clientes con contratos mensuales (Month-to-month) tienen una tasa de evasión notablemente alta, con un 42.7%. En contraste, los contratos de uno y dos años tienen tasas de churn muy bajas (11.3% y 2.8%, respectivamente), lo que demuestra que los contratos a largo plazo son un indicador de alta fidelidad.

Servicio de internet: El servicio de fibra óptica (Fiber optic) muestra una tasa de churn significativamente mayor (41.8%) en comparación con el servicio DSL (18.9%) y los clientes sin internet (7.2%). Esto podría indicar problemas de calidad de servicio, precios o insatisfacción específica con la fibra óptica.

Método de pago: El pago electrónico (Electronic check) está fuertemente asociado con una tasa de evasión del 45.1%. Los métodos de pago automáticos (tarjeta de crédito y transferencia bancaria) muestran tasas mucho más bajas, lo que sugiere que la automatización del pago reduce la fricción y el riesgo de churn.

Facturación: Los clientes que optan por la facturación sin papel (Facturacion_Sin_Papel) tienen más del doble de riesgo de churn (33.4%) que aquellos que reciben facturas en papel (16.3%).

Variables numéricas y Churn:

Antigüedad del cliente (Meses_Contrato): Los clientes que se dan de baja tienen una antigüedad promedio de solo 18 meses, significativamente menor que los clientes que se quedan (37 meses). Esto indica que los clientes nuevos son los que tienen mayor riesgo de irse.

Pago mensual (Pago_Mensual): Los clientes que se van tienen un pago mensual promedio más alto ($74.6) que los que se quedan ($61.4). Esto podría sugerir que los clientes con facturas más altas son más sensibles al precio o perciben un menor valor por su dinero.

Servicios adicionales: La falta de servicios de seguridad en línea, respaldo, protección de dispositivos y soporte técnico (Seguridad_Online, Backup_Online, Proteccion_Dispositivo, Soporte_Tecnico) está relacionada con una mayor propensión a la evasión. Esto sugiere que estos servicios crean un "efecto pegamento" que mantiene a los clientes.

🔹 Conclusiones e Insights
El análisis ha revelado que la evasión no es un fenómeno aleatorio, sino que está impulsado por factores específicos que pueden ser identificados y abordados. Los clientes con el mayor riesgo de churn son aquellos que:

Tienen contratos mensuales.

Utilizan el servicio de fibra óptica.

Pagan con cheques electrónicos.

Son clientes relativamente nuevos.

Tienen facturas mensuales altas.

No han contratado servicios de valor agregado como seguridad o soporte técnico.

No se encontró una correlación significativa entre el género del cliente y la tasa de evasión.

🔹 Recomendaciones
Basado en los hallazgos, se proponen las siguientes estrategias para reducir la evasión de clientes:

Fidelización de nuevos clientes: Implementar programas de bienvenida y seguimiento proactivo para clientes con menos de 12 meses de antigüedad. Ofrecer incentivos o promociones exclusivas para motivar la renovación de contratos anuales.

Transición a contratos a largo plazo: Promover activamente la migración de clientes con contratos mensuales a contratos anuales o bienales, ofreciendo descuentos o beneficios adicionales.

Optimización del servicio de fibra óptica: Investigar y mejorar la calidad del servicio de fibra óptica para abordar las posibles causas de insatisfacción. Esto podría incluir mejoras en la red, atención al cliente especializada o paquetes de servicio más competitivos.

Promoción de servicios de valor agregado: Incentivar la adopción de servicios como seguridad en línea y soporte técnico. Estos servicios no solo aumentan los ingresos, sino que también mejoran la experiencia del cliente y la retención.

Incentivar métodos de pago automáticos: Promover activamente el uso de tarjetas de crédito o transferencias bancarias automáticas, ofreciendo un pequeño descuento o beneficio a los clientes que migren de los cheques electrónicos.
